```python
Ind_Pak_ODI=ipl[((ipl['Team']=='India')&(ipl['Opponent']=='Pakistan'))|((ipl['Team']=='Pakistan')&(ipl['Opponent']=='India'))]
```

```python
Ind_Pak_ODI[['Team','Opponent','Result']]
```

```python
Ind_Pak_ODI[(Ind_Pak_ODI['Result']=='Win')]
```

```python
for Team,Opponent,Result in Ind_Pak_ODI[['Team','Opponent','Result']].values:
    print(f"{Team} vs {Opponent} : {Result}")
```

```python
Ind_Pak_ODI.query('Highest_Score > 100 | Run_scored > 300 ')
```

```python
Ind_Pak_ODI.rename(columns={"Highest Score":"Highest_Score","Run Scored":"Run_scored"},inplace=True)
```

```python
Ind_Pak_ODI
```

```python
Player_of_the_match_score_sum=Ind_Pak_ODI.groupby('Player Of The Match')['Highest_Score'].sum()

```

```python
player_of_the_match_df = Player_of_the_match_score_sum.reset_index()
player_of_the_match_df.columns = ['Player of the Match', 'Score sum']
player_of_the_match_df
```

```python
player_of_the_match_df.set_index('Player of the Match')['Score sum'].plot.pie(autopct='%1.1f%%', figsize=(10, 10), title='Player of the Match Score Sum')
```
